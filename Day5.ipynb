```python
#Part1 
input_data=open('input_D5.txt','r')
input_num=input_data.read().strip().split('\n')
input_data.close()
input_list=[num for num in input_num]
print(input_list[0])

line_parts=[]
lines=[]
for line in input_list:
    division_index= line.index('-')
    line_first_pair=line[:division_index -1]
    line_second_pair=line[division_index +3:]
    separator1=line_first_pair.index(',')
    x1=int(line_first_pair[:separator1])
    y1=int(line_first_pair[separator1+1:])
    separator2=line_second_pair.index(',')
    x2=int(line_second_pair[:separator2])
    y2=int(line_second_pair[separator2+1:])
    #print('(x,y):',(x1,y1),(x2,y2))
    if y1==y2:
        if x1<x2:
            for x in range(x1,x2+1):
                line_parts.append((x,y1))
        else:
            for x in range(x1,x2-1,-1):
                line_parts.append((x,y1))

    elif x1==x2:
        if y1<y2:
            for y in range(y1,y2+1):
                line_parts.append((x1,y))
        else:
            for y in range(y1,y2-1,-1):
                line_parts.append((x1,y)) 
    #print(line_parts)
#print(line_parts)
canvas=dict()
for i in line_parts:
    if i not in canvas:
        canvas[i]=1
    else:
        canvas[i]+=1 
#print(canvas)

count=0
for k,v in canvas.items():
    if v>1:
        count+=1

print(count)

```
